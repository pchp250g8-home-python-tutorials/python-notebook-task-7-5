При каких значениях x последовательность x, 10*x, x ^ lg(x) будет геометрической прогрессией?

```python
from sympy import Eq
from sympy import solve
from sympy import symbols
from sympy import log

x = symbols("x")
eq11 = Eq(x * x ** log(x,10), (10 * x) ** 2)
result = solve(eq11, x)
print(result)
```
